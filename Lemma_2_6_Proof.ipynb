In Lemma 2.6, we show that the bivariate quartic polynomial

$$p(x) = x_1^4 - x_1^2 + 2x_1 + x_2^4 - x_2^2 - 2x_2 + 2.4 - 2x_1x_2$$

is sos but cannot  be  written  as  the  sum  of  a  nonnegative  separable  and a nonnegative quadratic polynomial. Here, we show that $p(x) = z^T(x) Q z(x)$ where $z(x) = (1, x_1, x_2, x_1^2, x_1x_2, x_2^2)^T$ and $Q$ is a positive definitie matrix. We provide a rational $LDL^T$ factorization of $Q$ to show that it is positive definite.

We start by providing a Gram matrix. We have $p(x) = z^T(x) \left (\frac{1}{240}Q \right) z(x)$. 

Verify that $Q$ has integer entries (and so $\frac{1}{240} Q$ has rational entries).

```
import numpy as np
import sympy as sym
from numpy import linalg
from sympy.matrices import Matrix
from sympy import Symbol, pprint

x1 = Symbol('x1')
x2 = Symbol('x2')

p = x1**4 - x1**2 + 2*x1 + x2**4 - x2**2 - 2*x2 + 2.4 - 2*x1*x2;

z = Matrix([[1, x1, x2, x1**2, x1*x2, x2**2]])

Q = Matrix([[ 576,  240, -240, -205, -108, -205],
             [ 240,  170, -132,    0,  -99,  -99],
             [-240, -132,  170,   99,   99,    0],
             [-205,    0,   99,  240,    0,  -42],
             [-108,  -99,   99,    0,   84,    0],
             [-205,  -99,    0,  -42,    0,  240]])

display(Q)
```

Verify that $p(x) = z^T(x) \left (\frac{1}{240}Q \right) z(x)$.

```
sos_decomp = (1/240)*(z*Q*z.T).expand()
residual_in_sos_decomposition = sos_decomp[0] - p  # Should be an exact zero.
print('The residual in the sos decomposition of p(x) is', residual_in_sos_decomposition)
```

We now show that $Q$ is positive definite (and therefore the Gram matrix $\frac{1}{240} Q$ is positive definite). We show this by providing a rational $LDL^T$ decomposition of $Q$ where the smallest diagonal element of $D$ is strictly positive.

```
L, D = Q.LDLdecomposition()
```

Verify that the matrix $L$ has rational entries.

```
display(L)
```

Verify that the matrix $D$ has rational entries.

```
display(D)
```

Verify that $Q = LD L^T$.

```
Residual_in_LDL_factorization_of_Q = Q - L*D*L.T # Should be the zero matrix
display(Residual_in_LDL_factorization_of_Q)
```

Finally, we observe that the smallest diagonal element of $D$ is strictly positive.

```
print('The smallest diagonal element of D is: ', min(D.diagonal()))
```
